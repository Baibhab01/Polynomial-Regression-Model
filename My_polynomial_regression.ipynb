# Polynomial Regression

```python
#Importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
#imorting the dataset
dataset=pd.read_csv('Position_Salaries.csv')
X=dataset.iloc[:,1:-1]
y=dataset.iloc[:,-1]
```

```python
#Training the  Linear Regression model on the whole dataset
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(X,y)
```

```python
#Training the Polynomial Regression model on the whole dataset
from sklearn.preprocessing import PolynomialFeatures
poly_reg =PolynomialFeatures(degree=4)
X_poly=poly_reg.fit_transform(X)
lin_reg_2 =LinearRegression()
lin_reg_2.fit(X_poly,y)
```

```python
#Visualising the LR results
plt.scatter(X,y,color='red')
plt.plot(X,lin_reg.predict(X),color='blue')
plt.title('Truth or Bluff (Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
#Visualising the PLR rersults
plt.scatter(X,y,color='red')
plt.plot(X,lin_reg_2.predict(X_poly),color='blue')
plt.title('Polynomial Regression')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
##Using higher powers like n=4/5 instead of using n=2
### go back up and change the value of degree from 2 to 4/5
```

```python
#Predicting the new result with Linear regression
lin_reg.predict([[6.5]])
```

```python
#Predicting a new result with Polynomial regression
lin_reg_2.predict(poly_reg.fit_transform([[6.5]]))
```
